## Question 3: Wine Classification with Logistic Regression

### Setup

```python
random.seed(0)

wineData = io.loadmat("data/data_wine.mat")
print(wineData.keys())
print(wineData["X"].shape)
print(wineData["y"].shape)
print(wineData["X_test"].shape)

# Logistic function for arrays
def S(arr):
    return special.expit(arr)

# Logistic function for scalars
def s(num):
    return 1 / (1 + np.exp(-1 * num))

# Adds fictitious dimension(to represent alpha)
def addDimension(data):
    return np.concatenate((data, np.ones((data.shape[0], 1))), axis=1)

# Normalization function
def normalize(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    data = (data - mean) / std
    return data

# Cost function of logistic regression with mean cost and L2 regularization
def getCost(X, w, y, lambdaVal):
    meanCost = (-y.T @ np.log(S(X @ w))) - ((1 - y).T @ (np.log(1 - S(X @ w))))
    L2Norm = lambdaVal * (np.linalg.norm(w) ** 2)
    return meanCost + L2Norm

# Returns weight vector w after an update for batch gradient descent
def batchUpdate(X, w, y, lambdaVal, stepSize):
    gradient = -X.T @ (y - S(X @ w)) + (2 * lambdaVal * w)
    return w - stepSize * gradient

# Normalizing and adding fictitious dimension to training and test data
wineData["X"] = normalize(wineData["X"])
wineData["X"] = addDimension(wineData["X"])
wineData["X_test"] = normalize(wineData["X_test"])
wineData["X_test"] = addDimension(wineData["X_test"])

# Trains using batch gradient descent on dataset with initial weight vector w, lambda with value 
# lambdaVal, and step size(learning rate) with value stepSize. maxSize is the amount of iterations.
# def batchGDTrain(dataset, w, lambdaVal, stepSize, maxSize=5000, intervalSize=5):   
#     X = dataset["X"]
#     y = dataset["y"]
    
#     costs = []
#     for i in range(maxSize):
#         if i % intervalSize == 0: # Sample cost every 5 to save time
#             cost = getCost(X, w, y, lambdaVal)[0][0]
#             costs.append(cost)
#         w = batchUpdate(X, w, y, lambdaVal, stepSize)
#     return costs

# Returns weight vector w after an update for stochastic gradient descent
def stochasticUpdate(X_i, w, y_i, lambdaVal, stepSize):
    # numpy was being weird about X_i and y_i's shape
    X_i = np.reshape(X_i, (X_i.shape[0], 1))
    y_i = np.reshape(y_i, (1, 1))
    gradient = -X_i @ (y_i - s(X_i.T @ w)) + (2 * lambdaVal * w)
    return w - stepSize * gradient

# Trains using stochastic gradient descent
# def stochasticGDTrain(dataset, w, lambdaVal, stepSize, maxSize=10000, intervalSize=10):
#     X = dataset["X"]
#     y = dataset["y"]
    
#     costs = []
#     for i in range(maxSize):
#         if i % intervalSize == 0: # Sample cost every 5 to save time
#             cost = getCost(X, w, y, lambdaVal)[0][0]
#             costs.append(cost)
            
#         index = i % X.shape[0]
#         X_i = X[index]
#         y_i = y[index]
#         w = stochasticUpdate(X_i, w, y_i, lambdaVal, stepSize)
#     return costs

# Trains using stochastic gradient descent with a shrinking step size
# def shrinkingStochasticGDTrain(dataset, w, lambdaVal, delta, maxSize=10000, intervalSize=10):
#     X = dataset["X"]
#     y = dataset["y"]
    
#     costs = []
#     for i in range(maxSize):
#         if i % intervalSize == 0: # Sample cost every 5 to save time
#             cost = getCost(X, w, y, lambdaVal)[0][0]
#             costs.append(cost)
            
#         index = i % X.shape[0]
#         X_i = X[index]
#         y_i = y[index]
#         w = stochasticUpdate(X_i, w, y_i, lambdaVal, delta / (i + 1))
#     return costs

def results_to_csv(y_test, path):
    y_test = y_test.astype(int)
    df = pd.DataFrame({'Category': y_test})
    df.index += 1  # Ensures that the index starts at 1. 
    df.to_csv(path, index_label='Id')
```

### Part 2: Batch Gradient Descent Code

```python
lambdaVal = 0.25 # Optimized using cross-validation below
print("Regularization Parameter: {}".format(lambdaVal))
stepSize = 1/wineData["y"].shape[0] # Optimized using cross-validation below
print("Step Size: {}".format(stepSize))
w = np.ones((wineData["X"].shape[1], 1)) # Can be any nonzero vector

X = wineData["X"]
y = wineData["y"]

maxSize = 5000 # Number of iterations/updates
intervalSize = 5 # Interval between samples
costs = []

# Trains using batch gradient descent
for i in range(maxSize):
    if i % intervalSize == 0:
        cost = getCost(X, w, y, lambdaVal)[0][0]
        costs.append(cost)
    w = batchUpdate(X, w, y, lambdaVal, stepSize)
    
#costs = batchGDTrain(wineData, initialW, lambdaVal, stepSize, maxSize, intervalSize) # List of sampled costs
print(costs[-1])

plt.plot(range(0, maxSize, intervalSize), costs)
plt.xlabel("Dataset Size")
plt.ylabel("Cost")
plt.title("Batch Gradient Descent")
plt.show()
```

```python
# Using cross-validation for hyperparameter tuning and finding accuracy(for Kaggle submission)
dataSize = X.shape[0]
indices = np.arange(dataSize)
np.random.shuffle(indices)
shuffledData = X[indices]
shuffledLabels = y[indices]

maxSize = 5000 # Number of iterations/updates
intervalSize = maxSize // 1000 # Interval between samples

lambdaVal = 0.25
stepSize = 1/wineData["y"].shape[0] # Need to optimize this

k = 5 # Number of folds
foldSize = dataSize // k
w = np.ones((wineData["X"].shape[1], 1)) # Can be any nonzero vector
trainingAccuracy = 0
validationAccuracy = 0
for i in range(k):
    start = int(foldSize * i)
    end = int(foldSize * (i + 1))

    ithFoldTrainData = np.concatenate([shuffledData[:start], shuffledData[end:]])
    ithFoldTrainLabels = np.concatenate([shuffledLabels[:start], shuffledLabels[end:]])
    ithFoldValData = shuffledData[start:end]
    ithFoldValLabels = shuffledLabels[start:end]

    for j in range(maxSize):
        w = batchUpdate(ithFoldTrainData, w, ithFoldTrainLabels, lambdaVal, stepSize)

    trainingPredictions = (S(ithFoldTrainData @ w) >= 0.5).astype(int)
    validationPredictions = (S(ithFoldValData @ w) >= 0.5).astype(int)
    trainingAccuracy += np.sum(trainingPredictions == ithFoldTrainLabels) / ((k - 1) * dataSize)
    validationAccuracy += np.sum(validationPredictions == ithFoldValLabels) / dataSize
print("Training Accuracy for {}-Fold Cross-Validation: {}".format(k, trainingAccuracy))
print("Validation Accuracy for {}-Fold Cross-Validation: {}".format(k, validationAccuracy))
```

### Part 4: Stochastic Gradient Descent Code

```python
lambdaVal = 0.01 # Optimized using cross-validation below
print("Regularization Parameter: {}".format(lambdaVal))
stepSize = 1/wineData["y"].shape[0] # Optimized using cross-validation below
print("Step Size: {}".format(stepSize))
w = np.ones((wineData["X"].shape[1], 1)) # Can be any nonzero vector

X = wineData["X"]
y = wineData["y"]

maxSize = 100000 # Number of iterations/updates
intervalSize = maxSize // 1000 # Interval between samples
costs = []

# Trains using Stochastic Gradient Descent
for i in range(maxSize):
    if i % intervalSize == 0: # Sample cost every 5 to save time
        cost = getCost(X, w, y, lambdaVal)[0][0]
        costs.append(cost)

    index = i % X.shape[0]
    X_i = X[index]
    y_i = y[index]
    w = stochasticUpdate(X_i, w, y_i, lambdaVal, stepSize)

#costs = stochasticGDTrain(wineData, initialW, lambdaVal, stepSize, maxSize, intervalSize) # List of sampled costs
print(costs[-1])

plt.plot(range(0, maxSize, intervalSize), costs)
plt.xlabel("Dataset Size")
plt.ylabel("Cost")
plt.title("Stochastic Gradient Descent")
plt.show()
```

```python
# Using cross-validation for hyperparameter tuning and finding accuracy(for Kaggle submission)
dataSize = X.shape[0]
indices = np.arange(dataSize)
np.random.shuffle(indices)
shuffledData = X[indices]
shuffledLabels = y[indices]

k = 5 # Number of folds
foldSize = dataSize // k
w = np.ones((wineData["X"].shape[1], 1)) # Can be any nonzero vector

maxSize = 100000 # Number of iterations/updates
intervalSize = maxSize // 1000 # Interval between samples

lambdaVal = 0.01
stepSize = 1/wineData["y"].shape[0]

trainingAccuracy = 0
validationAccuracy = 0
for i in range(k):
    start = int(foldSize * i)
    end = int(foldSize * (i + 1))

    ithFoldTrainData = np.concatenate([shuffledData[:start], shuffledData[end:]])
    ithFoldTrainLabels = np.concatenate([shuffledLabels[:start], shuffledLabels[end:]])
    ithFoldValData = shuffledData[start:end]
    ithFoldValLabels = shuffledLabels[start:end]

    for j in range(maxSize):
        index = j % ithFoldTrainData.shape[0]
        X_i = ithFoldTrainData[index]
        y_i = ithFoldTrainLabels[index]
        w = stochasticUpdate(X_i, w, y_i, lambdaVal, stepSize)

    trainingPredictions = (S(ithFoldTrainData @ w) >= 0.5).astype(int)
    validationPredictions = (S(ithFoldValData @ w) >= 0.5).astype(int)
    trainingAccuracy += np.sum(trainingPredictions == ithFoldTrainLabels) / ((k - 1) * dataSize)
    validationAccuracy += np.sum(validationPredictions == ithFoldValLabels) / dataSize
print("Training Accuracy for {}-Fold Cross-Validation: {}".format(k, trainingAccuracy))
print("Validation Accuracy for {}-Fold Cross-Validation: {}".format(k, validationAccuracy))
```

Batch gradient descent converged much faster than stochastic gradient descent. In addition, the cost for batch gradient descent is much lower than stochastic gradient descent.

```python
lambdaVal = 0.05 # Need to optimize this
print("Regularization Parameter: {}".format(lambdaVal))
delta = 0.3 # Need to optimize this
print("Initial Step Size: {}".format(delta))
w = np.ones((wineData["X"].shape[1], 1)) # Can be any nonzero vector

X = wineData["X"]
y = wineData["y"]

maxSize = 100000 # Number of iterations/updates
intervalSize = maxSize // 1000 # Interval between samples
costs = []

# Trains using Stochastic Gradient Descent
for i in range(maxSize):
    if i % intervalSize == 0: # Sample cost every 5 to save time
        cost = getCost(X, w, y, lambdaVal)[0][0]
        costs.append(cost)

    index = i % X.shape[0]
    X_i = X[index]
    y_i = y[index]
    w = stochasticUpdate(X_i, w, y_i, lambdaVal, delta / (i + 1))


#costs = shrinkingStochasticGDTrain(wineData, initialW, lambdaVal, delta, maxSize, intervalSize) # List of sampled costs
print(costs[-1])

plt.plot(range(0, maxSize, intervalSize), costs)
plt.xlabel("Dataset Size")
plt.ylabel("Cost")
plt.title("Stochastic Gradient Descent with Shrinking Step Size")
plt.show()
```
